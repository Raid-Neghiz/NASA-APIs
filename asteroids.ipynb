```python
import json , requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start_date = "2024-12-22"
end_date = "2024-12-22"
url = "https://api.nasa.gov/neo/rest/v1/feed?start_date={}&end_date={}&api_key={}".format(start_date , end_date , api_key)
```

```python
requests = requests.get(url)
result = requests.json()
result 
```

```python
result = result['near_earth_objects']
result
```

```python
list_dates = []
for date in result.keys() : 
    list_dates.append(date)

list_dates
```

```python
data_list = []


for asteroid in range(len(result[list_dates[0]])):
    data = {
        'ID': result[list_dates[0]][asteroid]['id'],
        'Nom de l\'astéroïde': result[list_dates[0]][asteroid]['name'],
        'Diamètre minimal estimé (km)': float(result[list_dates[0]][asteroid]['estimated_diameter']['kilometers']['estimated_diameter_min']),
        'Magnitude absolue': float(result[list_dates[0]][asteroid]['absolute_magnitude_h']),
        'Vitesse relative (km/s)': float(result[list_dates[0]][asteroid]['close_approach_data'][0]['relative_velocity']['kilometers_per_second'])
    }
    data_list.append(data)

df = pd.DataFrame(data_list)
df
```

```python
df.to_csv('asteroides.csv', index=False, encoding='utf-8')
```
